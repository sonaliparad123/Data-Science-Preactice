```python
import pandas as pd
rating=pd.read_csv("ratings_small.csv")
rating
```

```python
movies=pd.read_csv("movies.csv")
movies
```

```python
movies_data=pd.merge(movies,rating)
movies_data
```

```python
gr=pd.DataFrame({'Mov':movies_data['title'],'Avg':movies_data['rating']})
gr
```

```python
ratings=gr.groupby(['Mov']).mean()
ratings
```

```python
rating_count=gr.groupby(['Mov']).count()
rating_count
```

```python
rating_mean_count = pd.DataFrame(ratings) 
```

```python
rating_mean_count
```

```python
rating_mean_count['rating_count']=rating_count
```

```python
#Recommandations System
import numpy as np
import pandas as pd
r_cols=['user_id','movieId','rating','unix_timestamp']
ratings=pd.read_csv("ratings_small.csv",names=r_cols,header=0)

#Getting number of users and moview from the dataset.
user_ids=ratings.user_id.unique().tolist()
movie_ids=ratings.movieId.unique().tolist()
print("Number of Users:{}".format(len(user_ids)))
print("Number of Users:{}".format(len(movie_ids)))
```

```python
user_movie_rating=movies_data.pivot_table(index='userId',columns='title',values='rating')
user_movie_rating.head()
```

```python
#we know tha each column contains all the users ratings for a particular movie Let's find all the user 
# ratings for the Forrest Gump (1994) and find the movies similar to it we choose this mpvies since it has the
# the highest no of ratings and we want to find the corelation between movies that have a higher no of ratings.
# To find the user ratings for Forrest Gump (1994) execute the 
forrest_gump_ratings=user_movie_rating['Forrest Gump (1994)']
forrest_gump_ratings.head(20)
```

```python
movies_like_forest_gump=user_movie_rating.corrwith(forrest_gump_ratings)
corr_forrest_gump=pd.DataFrame(movies_like_forest_gump,columns=["Correlation"])
corr_forrest_gump.dropna(inplace=True)
corr_forrest_gump.head(20)
```

```python
corr_forrest_gump.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_forrest_gump=corr_forrest_gump.join(rating_mean_count['rating_count'])
corr_forrest_gump.head()
```

```python
corr_forrest_gump[corr_forrest_gump['rating_count']>50].sort_values("Correlation",ascending=False).head(10)
```
