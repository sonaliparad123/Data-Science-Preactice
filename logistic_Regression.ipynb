```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("train.csv")
df
```

```python
sex=pd.get_dummies(df['Sex'])
embark=pd.get_dummies(df['Embarked'])
```

```python
embark
```

```python
df.head(5)
```

```python
a=df['Age'].mean()
df['Age'].fillna(a, inplace=True)
```

```python
df
```

```python
titanic_clean=pd.concat([df,sex,embark],axis=1)
titanic_clean
```

```python
titanic_clean.rename(columns={1:"female",0:"male"})
```

```python
titanic_clean.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
titanic_clean
```

```python
#logistic Regresion
x= titanic_clean[['Age','Pclass',0,1,'S','Q','C']]
y= titanic_clean['Survived']
```

```python
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

model_log = LogisticRegression()
model_log.fit(x_train,y_train)

y_pred=model_log.predict(x_test)
print("Test score:{:.6f}".format(np.mean(y_pred==y_test)))
```

```python
from sklearn import metrics
print("\n","Expected prediction  accuracy:",metrics.accuracy_score(y_test,y_pred))
```
