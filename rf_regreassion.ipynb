```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
insure=pd.read_csv("Boston.csv")
insure
```

```python
X=pd.DataFrame(insure[['rm','lstat','crim','tax','age']])
y=pd.DataFrame(insure['medv'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)#target 80-20
```

```python
dt_reg=RandomForestRegressor(n_estimators=1000,random_state=42)
model=dt_reg.fit(x_train,y_train)
```

```python
y_predict=dt_reg.predict(x_test)
```

```python
#Error estimation
#Error estimation
from sklearn import metrics
print("Mean Absolute Error:",metrics.mean_absolute_error(y_test,y_predict))
print("Mean Square Error:",metrics.mean_squared_error(y_test,y_predict))
print("Root Mean Square Error:",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
import pandas as pd
col_names=x_train.columns.tolist()
print(col_names)
feature_imp=pd.Series(dt_reg.feature_importances_,index=col_names).sort_values(ascending=False)
feature_imp
```
