```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("train.csv")
df
```

```python
sex=pd.get_dummies(df['Sex'])
embark=pd.get_dummies(df['Embarked'])
```

```python
embark
```

```python
a=df["Age"].mean()
df["Age"].fillna(a, inplace = True) 
  
```

```python
df
```

```python
titanic_clean=pd.concat([df,sex,embark],axis=1)
```

```python
titanic_clean
```

```python
titanic_clean.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
titanic_clean
```

```python
titanic_clean.rename(columns={1: "female", 0: "male"})
```

```python
y=titanic_clean['Survived']
X=titanic_clean[['Age','Pclass','male','female','C','Q','S']]
#X=titanic_clean[['Age',0,1,'C','Q','S']]
```
