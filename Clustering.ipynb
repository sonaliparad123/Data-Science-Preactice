```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled=scaler.fit_transform(milk.drop(['name'],axis=1))
milk_feat=pd.DataFrame(scaled,columns=milk.columns[1:])
milk_feat
```

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
data1=milk_feat[['water','protein','fat','lactose','ash']]
data1
```

```python
#create dendrogram
dendrogram=sch.dendrogram(sch.linkage(data1,method='ward'))
#create clusters
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(data1)
```

```python
#create dendrogram
dendrogram=sch.dendrogram(sch.linkage(data1,method='ward'))
#create clusters
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(data1)
```

```python
print(type(y_hc))
print(y_hc)
#using DataFrame.insert() to add a column
milk_feat.insert(5,"Group",y_hc,True)
print(milk_feat.head(20))
```

```python
#import KMeans
from sklearn.cluster import KMeans
# create kmeans object
kmeans=KMeans(n_clusters=4)
#fit kmeans object to data
#kmeans.fit(data)fit
data1=milk_feat[['water','protein','fat','lactose','ash']]
kmeans.fit(data1)
#print Location of custers Learned by kmeans object
print(kmeans.cluster_centers_)
#save new clusters for chart
y_km=kmeans.fit_predict(data1)
```
