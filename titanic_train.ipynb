```python
outliers=[]
def detect_outliers(df):
    threshold=3
    mean1=np.mean(df)
    std=np.std(df)
    
    for i in df:
        z_score=(i - mean1)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
outliers_pt=detect_outliers(df["Age"])
outliers_pt
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
#print(IQR)
LL=Q1-1.5*IQR
UL=Q3-1.5*IQR

titanic_Ageclean=df[((df['Age']>float(LL))&(df['Age']<float(UL)))]
```
