```python
bikeshare['date'] = pd.DatetimeIndex(bikeshare['datetime']).date
bikeshare['hour'] = pd.DatetimeIndex(bikeshare['datetime']).hour
bikeshare['month'] = pd.DatetimeIndex(bikeshare['datetime']).month
bikeshare['weekday'] = pd.DatetimeIndex(bikeshare['datetime']).strftime("%A")
```

```python
bikeshare
```

```python
bikeshare['holiday']=pd.Categorical(bikeshare.holiday)
bikeshare['workingday']=pd.Categorical(bikeshare.workingday)
bikeshare['weather']=pd.Categorical(bikeshare.weather)
bikeshare['season']=pd.Categorical(bikeshare.season)
```

```python
f,  (ax1)  =  plt.subplots(nrows=1, ncols=1, figsize=(13, 6))

ax1 = bikeshare[['month','count']].groupby(['month']).mean().reset_index().plot(kind='bar',
                                       legend = False, title ="Counts of Bike Rentals by Month", 
                                         stacked=True, fontsize=12, ax=ax1)
ax1.set_xlabel("Month", fontsize=12)
ax1.set_ylabel("Count", fontsize=12)
ax1.set_xticklabels(['January','February','March','April','May','June','July','August','September','October','November','December'])
```
